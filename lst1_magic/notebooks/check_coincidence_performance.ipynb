```python
import glob
import warnings
import numpy as np
import pandas as pd
from matplotlib import pyplot
```

```python
# --- ignore warnings ---
warnings.simplefilter('ignore')

# --- customize pyplot figure ---
pyplot.rcParams['figure.figsize'] = (12, 9)
pyplot.rcParams['font.size'] = 15
pyplot.rcParams['grid.linestyle'] = ':'

# --- get the default color cycle ---
color_cycle = pyplot.rcParams['axes.prop_cycle'].by_key()['color']
```

# Get the paths to DL1-coincidence data

```python
data_mask = '/home/yoshiki.ohtani/Crab/2020_11_18/2.event_coincidence/data/dl1_coincidence/*Run02923*.h5'
```

```python
file_paths = glob.glob(data_mask)
file_paths.sort()
```

# Check the coincidence profile

```python
path = file_paths[2]

print(f'Loading the following data:\n{path}')

df_features = pd.read_hdf(path, key='coincidence/features')
df_profile = pd.read_hdf(path, key='coincidence/profile')
```

```python
df_features
```

```python
df_profile
```

```python
pyplot.figure(figsize=(20, 7))
pyplot.xlabel('time offset [us]')
pyplot.ylabel('Number of coincident events')

sec2us = 1e6

for i_tel, tel_combo in enumerate(['lst1_m1', 'lst1_m2']):

    pyplot.plot(
        df_profile['offset'].values * sec2us, df_profile[f'n_coincidence_{tel_combo}'].values, 
        linestyle="None", marker='o', color=color_cycle[i_tel]
    )
    
    pyplot.step(
        df_profile['offset'].values * sec2us, df_profile[f'n_coincidence_btwn_{tel_combo}'].values, 
        where='post', color=color_cycle[i_tel], label=tel_combo
    )
    
    offset_avg = df_features.query(f'tel_combo == "{tel_combo}"')['offset_avg'].values * sec2us
    
    pyplot.plot(
        np.repeat(offset_avg, 2), (0, pyplot.ylim()[1]), linestyle='--', alpha=0.8, color=color_cycle[i_tel]
    )

pyplot.grid()
pyplot.legend()
```

# Check the coincidence performance

```python
data = pd.DataFrame()

print('Loading the input data...')

for i, path in enumerate(file_paths):
    
    if i%10 == 0:
        print(f'{i}/{len(file_paths)}')
    
    df = pd.read_hdf(path, key='coincidence/features')
    data = data.append(df)
    
print('Done.')
```

```python
data.head(10)
```

### averaged offsets

```python
pyplot.figure(figsize=(20, 7))
pyplot.xlabel('Unix time [sec]')
pyplot.ylabel('Averaged offset [us]')

for i_tel, tel_combo in enumerate(['lst1_m1', 'lst1_m2']):
    
    df = data.query(f'tel_combo == "{tel_combo}"')

    pyplot.plot(
        df['mean_time_unix'].values, df['offset_avg'].values * sec2us, 
        marker='o', markersize=4, linewidth=1, label=tel_combo
    )
    
pyplot.grid()
pyplot.legend()
```

### Number of coincident events

```python
pyplot.figure(figsize=(20, 7))
pyplot.xlabel('Unix time [sec]')
pyplot.ylabel('Number of coincident events')

for i_tel, tel_combo in enumerate(['lst1_m1', 'lst1_m2']):
    
    df = data.query(f'tel_combo == "{tel_combo}"')

    pyplot.plot(
        df['mean_time_unix'].values, df['n_coincidence'].values, 
        marker='o', markersize=4, linewidth=1, label=tel_combo
    )
    
pyplot.grid()
pyplot.legend()
```

### Ratio of coincident events over MAGIC events

```python
pyplot.figure(figsize=(20, 7))
pyplot.xlabel('Unix time [sec]')
pyplot.ylabel('Ratio (= N$_\mathrm{coincidence}$/N$_\mathrm{magic}$)')
pyplot.ylim(0, 1)

for i_tel, tel_combo in enumerate(['lst1_m1', 'lst1_m2']):
    
    df = data.query(f'tel_combo == "{tel_combo}"')

    pyplot.plot(
        df['mean_time_unix'].values, (df['n_coincidence']/df['n_magic']).values, 
        marker='o', markersize=4, linewidth=1, label=tel_combo
    )
    
pyplot.grid()
pyplot.legend()
```
